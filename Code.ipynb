## Greedy by Weight


```python
#greedy by weight
import time

profit = [7500, 25000, 35000, 15000, 50000, 75000, 3000, 25000, 22500,
          10000, 10000, 10000, 16000, 90000, 4000, 40000, 35000, 15000, 1500, 6000,
          14000, 44000, 20000, 20000, 4000, 4000, 20000, 5500, 1000, 7500, 5000,
          2500, 10000, 11000, 12500, 15000, 25000]

weight = [20, 10, 10, 10, 20, 10, 3, 10, 18, 45, 90, 25, 40, 20, 40, 10, 10, 15, 3,
           30, 20, 20, 20, 20, 20, 20, 20, 10, 10, 10, 10, 10, 5, 15, 10, 5,
           25]

M = 500 #batasan berat
n = len(profit) #jumlah barang
m = len(weight) #jumlah matriks
weight0 = weight
M0 = M
n0 = n
m0 = m
profit0 = profit
maksm = sum(weight)

if M >= maksm:
    print('Semua barang terangkut')
    untung = sum(profit)
    print(f'jumlah keuntungan= {untung}')
else:
    if n != m:
        print('Error, matrik profit dan weight tidak sama')
    else:
        # Greedy by Weight
        items = list(range(1, n + 1))  # Membuat list yang berisi indeks barang, mulai dari indeks 1
        items.sort(key=lambda i: weight[i - 1])  # Urutkan barang berdasarkan bobot
        berat = 0 #membuat variabel untuk menyimpan berat barang yang telah dipilih
        selected_items = []

        start_time = time.time()
        for i in items:
            if berat + weight[i - 1] <= M: #memeriksa apakah barang tsb bisa dimasukkan dgn syarat tidak lebih dr batas maks
                berat += weight[i - 1]  #jika bisa, maka berat yg dipilih akan ditambahkan dg berat barang tsb
                selected_items.append(i) #barang tsb ditambahkan ke dalam list barang

        end_time = time.time()
        execution_time = end_time - start_time

        print(f"Execution time: {execution_time} seconds \n")

        untung = sum(profit[i - 1] for i in selected_items) #menghitung total keuntungan dari setiap barang yg telah dipilih

        print('Greedy By Weight')
        print('Hasil')
        print(f'jenis barang yang diangkut: {selected_items}')
        print(f'jumlah beban yang diangkut: {berat}')
        print(f'jumlah keuntungan: {untung}')
```

## Greedy by Profit

```python
# Greedy by Profit
import time

profit = [7500, 25000, 35000, 15000, 50000, 75000, 3000, 25000, 22500,
          10000, 10000, 10000, 16000, 90000, 4000, 40000, 35000, 15000, 1500, 6000,
          14000, 44000, 20000, 20000, 4000, 4000, 20000, 5500, 1000, 7500, 5000,
          2500, 10000, 11000, 12500, 15000, 25000]

weight = [20, 10, 10, 10, 20, 10, 3, 10, 18, 45, 90, 25, 40, 20, 40, 10, 10, 15, 3,
           30, 20, 20, 20, 20, 20, 20, 20, 10, 10, 10, 10, 10, 5, 15, 10, 5,
           25]

M = 500 #batasan berat
n = len(profit) #jumlah barang
m = len(weight) #jumlah matriks
weight0 = weight #untuk menyimpan informasi berat

M0 = M
n0 = n
m0 = m
profit0 = profit
maksm = sum(weight)

items = list(range(1, n + 1))  # Membuat list yang berisi indeks barang, mulai dari indeks 1
items.sort(key=lambda i: profit[i - 1], reverse=True)  # Urutkan barang berdasarkan keuntungan (descending)
berat = 0 #membuat variabel untuk menyimpan berat barang yang telah dipilih
selected_items = []

start_time = time.time()
for i in items:
    if berat + weight[i - 1] <= M: #memeriksa apakah barang tsb bisa dimasukkan dgn syarat tidak lebih dr batas maks
        berat += weight[i - 1] #jika bisa, maka berat yg dipilih akan ditambahkan dg berat barang tsb
        selected_items.append(i) #barang tsb ditambahkan ke dalam list barang

end_time = time.time()
execution_time = end_time - start_time

print(f"Execution time: {execution_time} seconds \n")

untung = sum(profit[i - 1] for i in selected_items) #menghitung total keuntungan dari setiap barang yg telah dipilih

print('Greedy By Profit')
print('Hasil')
print(f'jenis barang yang diangkut: {selected_items}')
print(f'jumlah beban yang diangkut: {berat}')
print(f'jumlah keuntungan: {untung}')
```

## Greedy by Density

```python
#greedy by density
import numpy as np
import time

# Define variables
profit = [7500, 25000, 35000, 15000, 50000, 75000, 3000, 25000, 22500,
         10000, 10000, 10000, 16000, 90000, 4000, 40000, 35000, 15000, 1500, 6000,
         14000, 44000, 20000, 20000, 4000, 4000, 20000, 5500, 1000, 7500, 5000,
         2500, 10000, 11000, 12500, 15000, 25000]
weight = [20, 10, 10, 10, 20, 10, 3, 10, 18, 45, 90, 25, 40, 20, 40, 10, 10, 15, 3,
          30, 20, 20, 20, 20, 20, 20, 20, 10, 10, 10, 10, 10, 5, 15, 10, 5,
          25]

M = 500 # Batasan berat
n = len(profit) # Jumlah barang
m = len(weight) # Jumlah matriks
weight = np.array(weight)
profit = np.array(profit)

# Calculate profit-to-weight ratio
rasio = profit / weight

# Sort profit-to-weight ratio in descending order
R = np.sort(rasio)[::-1]

# Reconstruct the sorted profit and weight arrays
P = np.zeros(n)
W = np.zeros(n)
urutan = np.zeros(n)

start_time = time.time()
for i in range(n):
    a = R[i]
    for j in range(m):
        b = rasio[j]
        if a == b:
            P[i] = profit[j]
            W[i] = weight[j]
            urutan[i] = j
            rasio[j] = -b
            break

end_time = time.time()
execution_time = end_time - start_time

print(f"Execution time: {execution_time} seconds \n")

# Inisiasi variables
berat = 0
i = 0
untung = 0

# Determine the maximum weight that can be transported within the limit
while berat <= M:
    i += 1
    berat += W[i-1]

# Adjust the weight and calculate the profit
if berat > M:
    berat -= W[i-1]
    i -= 1
untung = np.sum(P[:i])

print('Greedy By Density')
print('Hasil')
print('Jenis barang yang di angkut:', ', '.join(map(str, urutan[:i].astype(int))))
print('Jumlah beban yang di angkut:', berat)
print('Jumlah keuntungan:', untung)
```

## Dynamic Programming

```python
import time

profit = [7500, 25000, 35000, 15000, 50000, 75000, 3000, 25000, 22500, 10000,
          10000, 10000, 16000, 90000, 4000, 40000, 35000, 15000, 1500, 6000, 14000,
          44000, 20000, 20000, 4000, 4000, 20000, 5500, 1000, 7500, 5000, 2500,
          10000, 11000, 12500, 15000, 25000]
weight = [20, 10, 10, 10, 20, 10, 3, 10, 18, 45, 90, 25, 40, 20, 40, 10, 10, 15, 3, 30,
          20, 20, 20, 20, 20, 20, 20, 10, 10, 10, 10, 10, 5, 15, 10, 5, 25]
M = 500

n = len(profit) #jumlah item/barang
m = M #kapasitas yang tersisa dr knapsack
p = M / m #rasio bobot maks terhadap kapasitas yang tersisa

z = [[0 for _ in range(m + 1)] for _ in range(n + 1)] # list 2D untuk menyimpan tabel dynamic programming

start_time = time.time() #perekaman waktu mulai eksekusi algoritma
for i in range(1, n + 1):
    y = [0] * (m + 1) #inisiasi list y yg merepresntasikan jmlh bobot yg tlh dipertimbangkan utk item ke-i
    z[i][0] = 0 #keuntungan maks utk knapsack [0] item pertama adalah 0
    for j in range(1, m + 1):
        y[j] = y[j - 1] + p #menghitung jumlah bobot yang telah dipertimbangkan utk kapasitas knapsack j
        if weight[i - 1] <= y[j]: #memeriksa apakah bobot item kr-i bisa muat ke knapsack j
            z[i][j] = max(z[i - 1][j], profit[i - 1] + z[i - 1][j - weight[i - 1]])
            #memilih keuntungan maks antara 2 kemungkinan (1) tanpa mengambil item ke i (2) ambil item ke i
        else:
            z[i][j] = z[i - 1][j]
            #kalau bobot item-i tdk muat, keuntungan maks ttp sama sprt tanpa waktu mengambil item ke i

end_time = time.time() #perekaman waktu mulai selesai algoritma
execution_time = end_time - start_time #menghitung waktu eksekusi algoritma

print(f"Execution time: {execution_time} seconds \n")

print("Dynamic Programming Table:")
for row in z:
    print(row)

#Loop utk cari solusi optimal
total_profit = z[n][m] #keuntungan maks yg dpt dicapai dgn mempertimbngkan semua item
total_weight = 0 #inisiasi total bobot item yg diambil

#Loop utk identifikasi item apasaja yg diambil
items_taken = [0] * (n + 1) #inisialisasi daftar utk itrm yg diambil/tidak
for i in range(n, 0, -1): #memeriksa  perbedaan keuntungan antara mengambil dan tdk mengambil item ke i
          #jika item diambil, menghitung kmbli kapasitas knapsack dan menandai item sebagai diambil (1)
          #jika item tdk diambil, maka item ditandai sebagai tdk diambil

    if z[i][m] != z[i - 1][m]:
        items_taken[i] = 1 #hapus elemen pertama yg tdk digunakan
        c = m * p
        q = c - weight[i - 1]
        m = round(q / p)
    else:
        items_taken[i] = 0

items_taken = items_taken[1:]
print("\nItems taken (1 means taken, 0 means not taken):", items_taken)

#menghitung total bobot item yang diambil
for i in range(n):
    if items_taken[i] == 1:
        total_weight += weight[i]

print("Total profit:", total_profit)
print("Total weight of items taken:", total_weight)
```
